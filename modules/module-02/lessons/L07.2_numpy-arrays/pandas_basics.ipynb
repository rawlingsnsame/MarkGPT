```python
# List comprehension

# List comprehensions let you build a new list by applying a transformation 
# or filter to an existing one, all in a single readable line.


numbers=[1, 2, 4, 5, 6, 9];

# traditional loop approach
square=[];

for n in numbers:
    square.append(n**2);
square
```

```python
# list comprehension... same results.. more allegant
squares=[n**2 for  n in numbers]
sqaures
```

```python
# Part 2: NumPy Arrays
import numpy as np
# import numpy
# NumPy (Numerical Python) is a library — not part of Python itself, but installable via pip install numpy. 
# At its heart, it provides a new data structure called 
# the ndarray (n-dimensional array) that is purpose-built for numerical computation.

# create a numpy array from a list
arr=np.array([1,2,4,5,6])
print(arr)

```

```python
print(type(arr))
```

```python
# The Critical Difference: Homogeneous and Contiguous
# This is the most important thing to understand about NumPy arrays. Unlike Python lists, a NumPy array:

# Must contain elements of the same type (homogeneous), and
# Stores those elements in a single contiguous block of memory — one right after the other, like soldiers standing in a line.

# Numpy enforces one type - it will try to "upcast" if you mix
arr=np.array([1, 2, 3.4]);
print(arr)

# We can explicitely specify the type
int_arr=np.array([1, 2, 3], dtype=np.int32);
print(int_arr)
```

```python
# Another example of upcasting
x=np.int32(10);
y=np.float64(5.5)
result=x+y;
print(result);
print(type(result))
```

## Vectorized Operations — The Superpower

With NumPy, mathematical operations apply to every element at once, with no Python loop needed:


```python
arr=np.array([1, 2,3,5,6]);
print(arr*2);
print(arr +10);
print(arr ** 2);
print(np.sqrt(arr))
```

```python
# Operations between  two arrays (element wise)

a=np.array([1,2,3])
b=np.array([12, 4,6]);
print(a+b)
```

## Broadcasting — A Subtle But Powerful Idea
Broadcasting is NumPy's ability to perform operations between arrays of different shapes, following a set of rules. It's an advanced topic, but worth planting in your mind early


```python
# Add a scalar to every element
```

```python
matrix=np.array([[1, 2, 4], [12, 23, 4]])
```

```python
print(matrix + 10)
```

NumPy automatically "broadcasts" the smaller array across the larger one. There's no equivalent to this in plain Python lists.

```python
row=np.array([10, 20, 30])
print(matrix + row)
```
