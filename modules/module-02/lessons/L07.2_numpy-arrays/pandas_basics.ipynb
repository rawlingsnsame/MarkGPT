```python
# Lists Are Mutable and Dynamic
# "Mutable" means you can change a list after creating it. "Dynamic" means it can grow or shrink freely.

numbers=[10, 20, 30];

numbers[1]=23;
numbers
```

```python
# Iterating over a list
scores=[84, 56, 34, 12];

for s in scores:
    print(s)
```

```python
# what if we need the index
for i, score in enumerate(scores):
    print(f"Student {i+1}: {score}")
```

```python
# List comprehension

# List comprehensions let you build a new list by applying a transformation 
# or filter to an existing one, all in a single readable line.


numbers=[1, 2, 4, 5, 6, 9];

# traditional loop approach
square=[];

for n in numbers:
    square.append(n**2);
square
```

```python
# list comprehension... same results.. more allegant
squares=[n**2 for  n in numbers]
sqaures
```

```python
squares
```

```python
s=[n+45 for n in numbers];
s
```

```python
# MANIPUATION PLUS A FILTER

even_squares=[n**2 for n in numbers if n%2==0];
even_squares
```
