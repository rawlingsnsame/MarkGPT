## Row vs Column Vectors

A row vector is shape (n,) or (1, n). A column vector is shape (n, 1). The distinction becomes critical in
matrix multiplication:

```python
v=np.array([1, 2, 3]) #shape (3, ) - ambigouse row or column
row=v.reshape(1, 3)  # Shape (1, 3) - explicitely a row vector
col=v.reshape(3,1) # Shape (3, 1) — explicitly a column vector
col
```

## 1.3 Vector Operations

To add two vectors, add each corresponding component. Vectors must have the same length

## 1.4 The Dot Product — The Most Import

Dot Product
Takes two vectors of the same length and produces a single scalar. Written as a · b or a^T
Formula from first principles: if a = [a1, a2, a3] and b = [b1, b2, b3], then:
a · b = a1×b1 + a2×b2 + a3×b3

```python
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])

# method 1=> explicit formula
manual=a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
print(manual)
```

```python
# cleaner method 2
np.dot(a,b)
```

```python
# method 3: @ operator (python 3>5+ prefered)
print(a@b)
```

### What does the dot product measure geometrically?
The dot product measures how much two vectors point in the same direction:

```python
a=np.array([1,0]) # Points right
b=np.array([0, 1]) # [Points up (perpendicular to a)]
c=np.array([1,0]) #Points right (same as a)
```

## VISUALLY 😎

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))

# Draw vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, label='a', color='blue')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, label='b', color='green')
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, label='c', color='red')

# Axes settings
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.axhline(0)
plt.axvline(0)
plt.grid()

plt.legend()
plt.title("Vector Diagram (a, b, c)")
plt.gca().set_aspect('equal')

plt.show()
```

```python
print(np.dot(a,b)) # 0.0 - Perpendicular vectors, dor product=0
print(np.dot(a,c)) #1.0 - Parallel vectors: dot product=maximum
print(np.dot(a, -c)) # -1.0 - Opposite direction, dot product=maximum
```

### 1.5 Vector Norm — The Length of a Vector

� Norm
The 'length' of a vector. The most common is the L2 norm (Euclidean length): ||v|| = sqrt(v1²
+ v2² + ... + vn²)

```python
v=np.array([3, 4])  # Classic 3-4-5 triangle
# manual calculations

manual=np.sqrt(v[0]**2 + v[1]**2)
```

```python
print(manual)
```

```python
# The Numpy way

print(np.linalg.norm(v))
```

```python
print(np.linalg.norm(v, ord=1)) # 7.0 (L1 NORM |3| + |4|=7)
```

```python
print(np.linalg.norm(v, ord=np.inf)) # Linf norm: max([3], [4])
```

## Unit Vectors — Normalizing

A unit vector has length exactly 1. You create one by dividing a vector by its norm. This strips away the
magnitude and keeps only direction:


```python
v=np.array([3.0, 4.0])

unit_v=v/np.linalg.norm(v)
print(unit_v)
```

```python
print(np.linalg.norm(unit_v))
```

#### � Deep Learning Connection
L2 regularization in neural networks penalizes large weight norms, preventing overfitting.
Batch normalization normalizes activations. Gradient clipping limits the gradient norm.
Norms are everywhere.
