## 1.4 The Dot Product — The Most Import

Dot Product
Takes two vectors of the same length and produces a single scalar. Written as a · b or a^T
Formula from first principles: if a = [a1, a2, a3] and b = [b1, b2, b3], then:
a · b = a1×b1 + a2×b2 + a3×b3

```python
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])

# method 1=> explicit formula
manual=a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
print(manual)
```

```python
# cleaner method 2
np.dot(a,b)
```

```python
# method 3: @ operator (python 3>5+ prefered)
print(a@b)
```

Takes two matrices A (m×k) and B (k×n) and produces C (m×n). The number of columns in

## 2.7 Broadcasting — NumPy's Superpower

� Broadcasting
NumPy's ability to perform operations on arrays of different shapes by automatically
'stretching' the smaller array to match the larger one, without copying data.

Broadcasting follows rules that are checked from the trailing dimensions. But for beginners, the most
common patterns are:

```python
A=np.array([[1,2,3], [4,3,5]]) # Pattern: Matrix + scalar (scalar stretches to fill matrix)
A
```

```python
A+10
```

```python
# pattern 2: Matrix + Row vector (row repeats for each row of matrix)
A=np.array([[1,3,4], [4,5,6]])

b=np.array([10, 20, 30]) # shape (3, )
print(A+b) # b is broadcast across both row
```

```python
# Pattern 3: Column + Row vector (outer addition)

col=np.array([[1], [2], [3]]) # shape (3, 1)
row=np.array([[10, 20, 30]]) # shape (3, )
print(col + row)
```

```python
# � Deep Learning Connection
# Adding biases to neuron outputs uses broadcasting: output = (batch @ W) + b, where
# output is (batch_size, neurons) and b is (neurons,). NumPy broadcasts b across every
# sample in the batch automatically.
```

## 3.1 What Is a Linear Transformation?
### � Linear Transformation

A function that maps vectors to other vectors using a matrix. When you multiply a matrix A
by a vector v, you are transforming (rotating, scaling, shearing, projecting) v into a new
vector.

```python
import numpy as np

# A 2x2 matrix transforms 2D vectors
# This matrix SCALES x BY 2 AND y by 3

A=np.array([[2,0], [0, 3]])
```

```python
A
```

```python
v=np.array([1,1])
v_transformed=A@v  # Apply the transformation
v_transformed
```

## Visualising the Transformation



```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def plot_transformation(A, title=''):
    # create grid of points
    xs=np.linspace(-2,2,10)
    # Generate 10 evenly spaced numbers between -2 and 2
    # added
    ys=np.linspace(-2, 2, 10)
    points=[]
    for x in xs:
        for y in ys:
            points.append([x,y]);
            
    points=np.array(points).T
    # convert everything to numpy array and then transpose to have x's as rows and y's as second row
    transformed=A@points
    fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(12, 5))
    ax1.scatter(points[0], points[1], color='navy', s=20)
    ax1.set_title('Before Transformation')
    ax1.set_xlim(-5, 5);
    ax1.set_ylim(-5, 5)
    ax1.axhline(0, color='gray', lw=0.5)
    ax1.axvline(0, color='gray', lw=0.5)
    ax1.grid(True, alpha=0.3)
    ax1.set_aspect('equal')

    ax2.scatter(transformed[0], transformed[1], color='crimson', s=20)
    ax2.set_title(f'After: {title}')
    ax2.set_xlim(-5, 5);
    ax2.set_ylim(-5, 5);
    ax2.axhline(0, color='gray', lw=0.5)
    ax2.axvline(0, color='gray', lw=0.5)
    ax2.grid(True, alpha=0.3)
    ax2.set_aspect('equal')
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np

A=np.array([[2,0], [0, 2.5]])
plot_transformation(A, 'scale')
```
