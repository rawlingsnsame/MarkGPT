```python
import numpy as np
import matplotlib .pyplot as plt
```

```python
players = {
 'Messi': np.array([90, 75, 82, 70, 95, 30], dtype=float),
 'Ronaldo': np.array([88, 60, 90, 85, 80, 35], dtype=float),
 'Kante': np.array([20, 40, 88, 98, 85, 97], dtype=float),
 'Mbappe': np.array([78, 55, 99, 88, 80, 38], dtype=float),
 'De Bruyne':np.array([50, 88, 75, 82, 97, 45], dtype=float),
 'Van Dijk': np.array([15, 20, 78, 90, 78, 95], dtype=float),
 'Salah': np.array([75, 60, 91, 80, 82, 42], dtype=float),
 'Modric': np.array([30, 70, 72, 90, 96, 80], dtype=float),
 'Lewandowski': np.array([92, 50, 78, 80, 82, 28], dtype=float),
 'Alisson': np.array([5, 5, 65, 85, 88, 70], dtype=float),
}
players
```

```python
# store names in a consistent order for matrix operations later
names=list(players.keys())
names
```

```python
# stack vectors in a matrix: shape (10, 6) - 10 players, 6 stats each

P=np.array([players[n] for n in names])
P
```

## Step 2 — Computing vector norms

Before we compute similarity, let us explicitly compute and display the norms for each player. This
reinforces the mathematical idea that norm = 'how long is this vector', which for our stats means
'how large are this player's numbers overall'. A goalkeeper will naturally have a smaller norm than a
forward because five of the six stats are lower

```python
l2_norms=np.linalg.norm(P, axis=1) # shape (10, )
l1_norms=np.linalg.norm(P, axis=1, ord=1)

print("\n Norms for each player: ");
print(f'{"player":<15}  {"L2 Norm":>10}  {"L1 Norm":>10}')
print('-' * 42)

for name, l2, l1 in zip(names, l2_norms, l1_norms):
    print(f'{name:<15} {l2:>10.2f} {l1:>10.2f}')
```

```python
# The 'most balanced' player is the one whose norm is closest to having
# equal components. We can find this by comparing max(v) / min(v) ratio.
# A perfectly balanced player has ratio = 1 (all stats equal).

balance_ratio=[]

for name in names:
    v=players[name]
    ratio=np.max(v)/ (np.min(v) + 1e-8) # 1e-8 avoid divison by zero
    balance_ratio.append(ratio)

most_balanced_idx=np.argmin(balance_ratio)
print(f'\n Most balanced player: {names[most_balanced_idx]}')
print(f'Max/Min ratio: {balance_ratio[most_balanced_idx]:.2f}')
print('(A ratio of 1.0 would mean perfectly equal in every stat)')
```

## Step 3 — Cosine similarity, built from scratch

This is the core of the project. We implement cosine similarity entirely from first principles — no
sklearn, no scipy. 

```python
def cosine_similarity(a, b):
    dot=np.dot(a,b);
    norm_a=np.linalg.norm(a);
    norm_b=np.linalg.norm(b);
    return dot/(norm_a*norm_b);
cosine_similarity(9,4)
```

```python
def find_similar_players(target_name, players_dict, top_k=3):
    target_vec=players_dict[target_name]
    similarities=[];

    for name, vec in players_dict.items():
        if name==target_name:
            continue # skip self-comparisons
        sim=cosine_similarity(target_vec, vec)
        similarities.append((name, sim))

    # sort by simiarity, dewscending (most similar first)
    similarities.sort(key=lambda x: x[1], reverse=True)
    print(f'\n Top {top_k} players most similar to {target_name}: ')
    for rank, (name, sim) in enumerate(similarities[:top_k], 1):
        print(f'{rank}. {name:<15} cosine similarity= {sim:.4f} ')
# Test it
find_similar_players('Mbappe', players)
find_similar_players('Kante', players)
find_similar_players('Messi', players)


```

## Step 4 — Full similarity matrix

It is instructive to build the entire N x N similarity matrix, where entry [i, j] is the cosine similarity
between player i and player j. This matrix is symmetric (sim(a,b) = sim(b,a)) and has 1.0 on the
diagonal (each player is perfectly similar to themselves). Building it efficiently uses matrix
operations rather than nested loops.

```python
def cosine_Similarity(P):
    # The cosine sim, for all pairs at once can be calculated as :
    # C=P_normalised @ P_normalised.T
    # compute the norm of each row (each player's vector)
    norms=np.linalg.norm(P, axis=1, keepdims=True) #  shapes (10, 1)
    # keepdims=True : keeps dimension to (10, 1) so we can easily do broadcasting
    P_norm=P/norms # brodcasting (10, 6) / (10, 1)=(10, 6)
    C=P_norm @ P_norm.T
    return C
C=cosine_Similarity(P);
print('Similarity matrix shape: ', C.shape) # (10, 10)
print('Diagonal (self-similarity): ', np.round(np.diag(C), 4))

    
```
